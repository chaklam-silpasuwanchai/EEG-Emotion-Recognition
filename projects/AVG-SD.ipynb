# AVG and SD calculator

```python
import pandas as pd
```

```python
import os

rootdir = "./output"

for subdir, dirs, files in os.walk(rootdir):
    for file in files:
        #print os.path.join(subdir, file)
        filepath = subdir + os.sep + file
        print(subdir)
        
        if filepath.endswith(".csv"):
            print (file)
            
            df = pd.read_csv(filepath, sep='|', header=0)

            print("AVG test_acc : ", df['test_acc'].mean())
            print("SD  test_acc : ", df['test_acc'].std())
            
            print("-"*50)
```
